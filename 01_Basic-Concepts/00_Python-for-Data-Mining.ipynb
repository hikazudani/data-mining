### Repeat While Condition


```python
# While loop: runs while the condition is True
print("=== COUNTING WITH WHILE ===")
counter = 1
while counter <= 5:
    print(f"Counter: {counter}")
    counter += 1  # Increment
```

### Flow Control
* break - Stops the loop completely
* continue - Skips to the next iteration
* Use Cases:
* break: Find a specific item, exit infinite loops
* continue: Skip invalid values, filter data


```python
# break - Completely stops the loop
for i in range(10):
    if i == 5:
        break  # Stops the loop here
    print(i)  # Prints 0, 1, 2, 3, 4
```

```python
# continue - Skips to the next iteration
# skip invalid values, filter data
for i in range(5):
    if i == 2:
        continue  # Skips when i = 2
    print(i)  # Prints 0, 1, 3, 4
```

### Functions: Basic Syntax
* Components:
* def: keyword to define a function
* largura, altura: parameters
* return: returned value


```python
def calculate_rectangle_area(width, height):
    """Calculates the area of a rectangle."""
    area = width * height
    return area

# Using the function
area1 = calculate_rectangle_area(5, 3)
area2 = calculate_rectangle_area(8, 4)
area3 = calculate_rectangle_area(10, 2)

print(f"Area 1: {area1}")  # Area 1: 15
print(f"Area 2: {area2}")  # Area 2: 32
print(f"Area 3: {area3}")  # Area 3: 20
```

```python
def function_name(parameter1, parameter2):
    """
    Docstring: description of the function
    Args: expected parameters
    Returns: what the function returns
    """
    # Function body
    result = parameter1 + parameter2
    return result

# Function call
value = function_name(10, 20)  # arguments
print(value)  # 30
```

### Lists: Ordered Collections of Data
* A list in Python is a data structure that allows storing a collection of items in a single variable, which can be of different types (numbers, strings, even other lists).
* Lists are mutable: it is possible to change values, add, and remove elements after creation.
* They support order: each element in the list has a defined position (index), starting at 0; the order is always preserved.
* They allow duplicate elements, that is, they can contain items with the same value.

* Syntax: they are defined between brackets, for example: [10, 'Python', 3.5].
* They support various operations like iterating with loops (for), slicing (list[1:3]), methods like append(), remove(), sort(), among others.
* They accept nesting: lists can contain other lists as elements, forming more complex structures (e.g., matrices/2D).
* They are widely used to organize, process, filter, sort, and manipulate collections of data in Python.


```python
# Creating lists
numbers = [1, 2, 3, 4, 5]
names = ["Ana", "Bruno", "Carlos"]
mixed = [1, "Python", 3.14, True]
empty_list = []

# Accessing elements (index starts at 0)
print(numbers[0])  # 1 (first element)
print(names[1])  # Bruno (second element)
print(numbers[-1])  # 5 (last element)
print(names[-2])  # Bruno (second-to-last element)

# List information
print(len(numbers))  # 5 (list length)
```


---
### Manipulating Lists

```python
fruits = ["apple", "banana", "orange"]

# Adding elements
fruits.append("grape")  # Adds to the end
print(fruits)  # ['apple', 'banana', 'orange', 'grape']

fruits.insert(1, "mango")  # Adds at position 1
print(fruits)  # ['apple', 'mango', 'banana', 'orange', 'grape']
```

```python
# Removing elements
# ['apple', 'mango', 'banana', 'orange', 'grape']
fruits.remove("banana")  # Removes by value
print(fruits)  # ['apple', 'mango', 'orange', 'grape']

removed_fruit = fruits.pop()  # Removes the last element and returns it
print(removed_fruit)  # grape

# Modifying elements
fruits[0] = "pear"
print(fruits)  # ['pear', 'mango', 'orange']
```

### Tuples: Immutable Lists
* A tuple is a data structure in Python used to group multiple elements in a single variable, just like lists.
* The main characteristic of tuples is immutability: once created, its elements cannot be modified, added, or removed.
* The size of the tuple also does not change.
* They are defined using parentheses: for example, (10, 'Python', 3.5).
* Tuples can store different types of data and can be nested (tuples within tuples).
* They allow access by index (position), and support operations like slicing and iteration in loops,
* exactly like lists.
* Ideal for representing information that should not be changed, offering security and better performance in some operations due to immutability.
* They are frequently used for multiple returns from functions, grouping fixed data, and other contexts where integrity is desired.


```python
# Creating tuples
coordinates = (10, 20)
rgb_colors = (255, 128, 0)
person_data = ("Ana", 25, "Engineer")

# Tuple with a single element (comma is required!)
single_tuple = (42,)  # Correct
single_tuple = (42)   # This is just a number!

# Accessing elements (same as lists)
print(coordinates[0])  # 10
print(person_data[-1])  # Engineer

# Attempting to modify a tuple causes an error
# coordinates[0] = 15  # TypeError
```

### Hashtables
* A hashtable (or hash table) is a data structure that stores key-value pairs.
* To quickly store and locate values, the hashtable uses a hash function: it transforms the key (for example, a name, number, or string) into an integer, called an index.
* This index indicates the position in the array (table) where the value will be stored or retrieved.
* It is very efficient: operations like insert, search, and remove are fast.
* Examples of use: storing people's ages by name, counting word frequency, implementing dictionaries in programming languages.

---
### Dictionaries: Organized Data

#### Keys must be unique and immutable (strings, numbers, tuples).

* A dictionary is an unordered collection of key: value pairs, where each key is unique.
* It can be created using {} or the dict() function, for example: d = {'name': 'Alice', 'age': 25}.
* Keys can be of immutable types (strings, numbers, tuples), and values can be of any type.
* Allows fast access to values from keys, without depending on order or indices like in lists.
* Dictionaries are mutable, that is, it is possible to add, modify, or remove pairs.


```python
# Creating dictionaries
person = {
    "name": "Maria",
    "age": 28,
    "profession": "Data Scientist",
    "active": True
}

# Accessing values
print(person["name"])  # Maria
print(person.get("age"))  # 28
```

```python
# Modifying and adding
person["age"] = 29  # Modify value
person["city"] = "São Paulo"  # Add new key

print(person)
```

```python
def analyze_survival(data):
    """
    Analyzes the Titanic survival rate.

    Args:
        data (list): List of dictionaries containing passenger data.
                     Each dictionary must have the key 'Survived' (0 or 1).

    Returns:
        dict: Dictionary with the following keys:
              - 'total' (int): Total number of passengers
              - 'survivors' (int): Number of survivors
              - 'rate' (float): Survival rate in %

    Example:
        >>> analyze_survival([{'Survived': 1}, {'Survived': 0}])
        {'total': 2, 'survivors': 1, 'rate': 50.0}
    """

    # Calculate total number of passengers
    total = len(data)

    # Count survivors using a generator expression
    # Equivalent to LINQ Count(p => p.Survived == 1) in C#
    survivors = sum(1 for passenger in data if passenger['Survived'] == 1)

    # Calculate survival rate as a percentage
    rate = round((survivors / total) * 100, 2)

    # Return dictionary with results
    return {
        'total': total,
        'survivors': survivors,
        'rate': rate
    }
```
